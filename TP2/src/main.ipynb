5. ¿Qué algoritmos se utilizan para resolver problemas modelados con lógica proposicional?

        Los algoritmos que se utilizan para resolver problemas modelados con lógica proposicional son los algoritmos de resolución. Estos algoritmos se utilizan para encontrar una secuencia de pasos que permitan llegar a una solución. Algunos de los algoritmos de resolución más comunes son: resolución de refutación, resolución de unificación y resolución de sustitución.

8. Describa cómo los planificadores hacen uso de algoritmos de búsqueda globales para encontrar una secuencia óptima de pasos a seguir para resolver un problema
    
            Los planificadores hacen uso de algoritmos de búsqueda globales para encontrar una secuencia óptima de pasos a seguir para resolver un problema. Estos algoritmos de búsqueda globales se utilizan para encontrar una secuencia de pasos que permitan llegar a una solución óptima. Algunos de los algoritmos de búsqueda globales más comunes son: búsqueda en profundidad, búsqueda en anchura, búsqueda de costo uniforme, búsqueda de A*, búsqueda de haz y búsqueda de escalada. Estos algoritmos de búsqueda globales se utilizan para encontrar una secuencia de pasos que permitan llegar a una solución óptima en un espacio de estados.
